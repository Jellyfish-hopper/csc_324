```python
import seaborn as sns
import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
```

```python
filename = "exam_results.csv"
score_df = pd.read_csv(filename)
score_df.head()
```

```python
score_df.rename(columns={"race/ethnicity":"ethinicity", "parental level of education":
                        "parent_education","math score": "math", 
                        "reading score":"reading",
                        "writing score":"writing",
                        "test preparation course":"pre"}, inplace=True)
score_df.head()
            
```

```python
score_df.dtypes
```

```python
f, (ax1, ax2, ax3) = plt.subplots(
    1, 3, figsize=(12, 4), sharey=True)
sns.barplot(x='gender', y='reading', data=score_df, ax=ax1)
sns.barplot(x='ethinicity', y='reading', data=score_df, ax=ax2)
sns.swarmplot(x='pre', y='reading', data=score_df, ax=ax3)
```

```python
f, (ax1, ax2, ax3) = plt.subplots(
    1, 3, figsize=(12, 4), sharey=True)
sns.barplot(x='gender', y='math', data=score_df, ax=ax1)
sns.barplot(x='ethinicity', y='math', data=score_df, ax=ax2)
sns.swarmplot(x='pre', y='math', data=score_df, ax=ax3)
```
